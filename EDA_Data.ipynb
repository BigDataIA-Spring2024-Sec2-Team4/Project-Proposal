```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('netflix_titles.csv')
df2 = pd.read_csv('amazon_prime_titles.csv')
df3 = pd.read_csv('hulu_titles.csv')
df4 = pd.read_csv('disney_plus_titles.csv')
```

```python
# Generate a word cloud image
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(df1['listed_in'].value_counts())

# Display the generated image:
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Genre Distribution of Netflix')
plt.show()
```
